## Tensor Operations on GPU

```python
torch.cuda.is_available()
```

```python
dev=torch.device("cuda")
```

```python
a=torch.tensor([1,2,3],device=dev)
```

```python
a
```

```python
#cpu vs GPU
import time
size=10000
cpu1=torch.rand(size,size)
cpu2=torch.rand(size,size)

start=time.time()
torch.matmul(cpu1,cpu2)
last=time.time()
print(f"cpu time : {last-start:.2f}")

cpu1=torch.rand(size,size,device=dev)
cpu2=torch.rand(size,size,device=dev)

start=time.time()
torch.matmul(cpu1,cpu2)
last=time.time()
print(f"gpu time : {last-start:.2f}")
```
