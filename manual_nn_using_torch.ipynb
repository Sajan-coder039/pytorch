```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv(path+"/breast-cancer.csv")
```

```python
df.drop("id",inplace=True,axis=1)
```

```python
X=df.drop("diagnosis",axis=1)
y=df["diagnosis"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#scaling
from sklearn.preprocessing import StandardScaler,LabelEncoder
scaler=StandardScaler()
x_train_scal=scaler.fit_transform(x_train)
x_test_scal=scaler.transform(x_test)
label=LabelEncoder()
y_train_scal=label.fit_transform(y_train)
y_test_scal=label.transform(y_test)
```

```python
###numpy to tensor
import torch
x_train_torch=torch.from_numpy(x_train_scal)
x_test_torch=torch.from_numpy(x_test_scal)
y_train_torch=torch.from_numpy(y_train_scal)
y_test_torch=torch.from_numpy(y_test_scal)
```

```python
x_train_torch.shape
```

```python
class MysimpleNN:
    def __init__(self,X):
        self.weights=torch.rand(X.shape[1],1,dtype=torch.float64,requires_grad=True)
        self.bias=torch.zeros(1,dtype=torch.float64,requires_grad=True)

    def forward(self,x):
        z=torch.matmul(x,self.weights)+self.bias
        return torch.sigmoid(z)
    def loss_cal(self,y_pred,y):
        epsilon=1e-7
        y_pred=torch.clamp(y_pred,epsilon,1-epsilon)
        loss=-((y*torch.log(y_pred))+((1-y)*torch.log(1-y_pred)))
        return loss.mean()
```

```python
learning_rate=0.12
epoch=525
```

```python
##create model
model=MysimpleNN(x_train_torch)
#forward propogation
for i in range(epoch):
    y_pred=model.forward(x_train_torch)
    #loss calculation
    loss=model.loss_cal(y_pred,y_train_torch)
    #backward
    loss.backward()
    #parameter upgrade
    with torch.no_grad():
        model.weights-=(learning_rate*model.weights.grad)
        model.bias-=(learning_rate*model.bias.grad)

    model.weights.grad.zero_()
    model.bias.grad.zero_()
    print(f"epoch: {i+1} and loss = {loss.item()}")
```

```python
with torch.no_grad():
    y_pred=model.forward(x_test_torch)
    y_pred=(y_pred>0.7).float()
    print(accuracy_score(y_pred,y_test_torch))
```
